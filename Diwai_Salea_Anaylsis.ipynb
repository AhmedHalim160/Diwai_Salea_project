```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Amount']=df['Amount'].astype(int)
```

```python
df['Amount'].dtype
```

```python
df.rename(columns={'Marital_Status':'Married'},inplace=True)
```

```python
df[['Age','Amount','Orders']].describe()
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
data=df.groupby(['Gender'],as_index=False)['Amount'].sum()
```

```python
sns.barplot(x='Gender',y='Amount',data=data)
plt.show()
```

## Frow above graphs we can see that the most buyers is females grater than men

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender' )
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df1=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Age Group',y='Amount',data=df1,palette='Set2')
plt.show()
```

```python
df1=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
plt.figure(figsize=(17,5))
sns.barplot(x='State',y='Orders',data=df1,palette='Set1')
plt.show()
```

## from above graphs we can see that the most of Orders and TotalSales are from Uttar pradesh,Maharashtra and karnatake

```python
ax=sns.countplot(data=df,x='Married',palette='deep')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
df1=df.groupby(['Married','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(9,8))
sns.barplot(x='Married',y='Amount',hue='Gender',data=df1)
```

## Frow above graphs we can see that the most of the buyers are not married (Women)

```python
plt.figure(figsize=(20,8))
ax=sns.countplot(data=df,x='Occupation',palette='colorblind')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
df1=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(25,4))
sns.barplot(x='Occupation',y='Amount',data=df1,palette='dark')
```

## from above graphs we can see that the most of buyers are working in ITSector,Healthcare and Aviation 

```python
df1=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(25,4))
sns.barplot(x='Product_Category',y='Amount',data=df1,palette='flare')
```

## from above graphs we can see that the most of the sold products are from Food, Clothing  and Electronics
